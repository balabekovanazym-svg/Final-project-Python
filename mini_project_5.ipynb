```python
class Users:
    def __init__(self, first_name, second_name, email, birth_year):
        self.first_name = first_name
        self.second_name = second_name
        self.email = email
        self.birth_year = birth_year
    def getAge(self):
        return 2023 - int(self.birth_year)
```

```python
test = Users('Magzhan','Kairanbay','magzhan.kairanbay@gmail.com', 1991)
```

```python
print(test.getAge())
```

```python
users_list = []
with open('user.txt','r') as file:
    for line in file:
        line = line.rstrip('\n')
        parts = line.split(',')
        user = Users(parts[1], parts[2], parts[3], parts[4])
        users_list.append(user)
```

```python
for user in users_list:
    print(user.first_name, user.second_name, user.email, user.birth_year, user.getAge())
```

```python
def filter(users, low=30, high=50):
    for user in users:
        age = user.getAge()
        if low < age < high:
            print(f'{user.email}, {age}')
```

```python
filter(users_list)
```
